```python
import tensorflow as tf
from tensorflow.keras import layers, models

# ১. ফোল্ডার থেকে ডেটা লোড করা এবং প্রি-প্রসেসিং করা
data_dir = '/content/drive/MyDrive/Rice_Disease_Project'

# ইমেজ সাইজ এবং ব্যাচ সাইজ সেট করা
img_height = 224
img_width = 224
batch_size = 32

# ডেটাসেট তৈরি (এখানেই ইমেজ রিসাইজ এবং রিস্কেল হচ্ছে)
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    image_size=(img_height, img_width),
    batch_size=batch_size,
    shuffle=True
)

# ২. পিক্সেল ভ্যালু নরমাল করা (0-255 কে 0-1 এর মধ্যে আনা)
normalization_layer = layers.Rescaling(1./255)
train_ds = train_ds.map(lambda x, y: (normalization_layer(x), y))

# প্রি-প্রসেসিং শেষে ডেটা চেক করা
for image_batch, labels_batch in train_ds.take(1):
    print("Image batch shape:", image_batch.shape)
    print("Label batch shape:", labels_batch.shape)
    break
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# মডেল আর্কিটেকচার তৈরি
model = models.Sequential([
    # প্রথম কনভল্যুশনাল লেয়ার
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(224, 224, 3)),
    layers.MaxPooling2D((2, 2)),

    # দ্বিতীয় কনভল্যুশনাল লেয়ার
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),

    # ফ্ল্যাটেন এবং ডেন্স লেয়ার
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(3, activation='softmax') # আপনার ৩টি রোগ তাই এখানে 3
])

# মডেল কম্পাইল করা
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# ট্রেনিং শুরু করা
history = model.fit(
    train_ds,
    epochs=10 # আপনি চাইলে ইপক সংখ্যা বাড়াতে পারেন
)
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models

# Traditional ML & Ensemble Models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBClassifier
from sklearn.ensemble import VotingClassifier, StackingClassifier
from sklearn.metrics import accuracy_score
```

```python
# ১. ডেটা অগমেন্টেশন লেয়ার (ফ্লোচার্টের Feature Engineering ধাপ)
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.1),
])

# ২. প্রি-ট্রেইনড বেস মডেল (ফিচার এক্সট্রাক্ট করার জন্য)
base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model.trainable = False

# ৩. ফিচার এক্সট্রাক্টর মডেল তৈরি
feature_extractor = models.Sequential([
    layers.Input(shape=(224, 224, 3)),
    data_augmentation,
    base_model,
    layers.GlobalAveragePooling2D() # এটি ছবিকে ১টি ফ্ল্যাট ভেক্টরে রূপান্তর করবে
])

print("Feature extractor model is ready!")
```

```python
# train_ds থেকে সব ছবি এবং লেবেল বের করা
images_list = []
labels_list = []

for images, labels in train_ds:
    images_list.append(images.numpy())
    labels_list.append(labels.numpy())

# সব ব্যাচ একসাথে জোড়া দেওয়া
X_raw = np.concatenate(images_list, axis=0)
y = np.concatenate(labels_list, axis=0)

# ফ্লোচার্ট অনুযায়ী ফিচার এক্সট্রাক্ট করা (Apply Model Training এর পূর্ববর্তী ধাপ)
print("Extracting features from images using MobileNetV2... (It may take a while)")
X_features = feature_extractor.predict(X_raw)

print(f"Features shape: {X_features.shape}")
print(f"Labels shape: {y.shape}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_features, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Training data size: {X_train.shape[0]}")
print(f"Testing data size: {X_test.shape[0]}")
```

```python
# ১. বেসলাইন ক্লাসিফায়ারসমূহ
lr = LogisticRegression(max_iter=1000, random_state=42)
dt = DecisionTreeClassifier(random_state=42)
svm = SVC(probability=True, random_state=42) # Soft Voting এর জন্য probability=True আবশ্যক
knn = KNeighborsClassifier()
gnb = GaussianNB()

# ২. এক্সজি বুস্ট ক্লাসিফায়ার
xgb = XGBClassifier(random_state=42, eval_metric='mlogloss')

# এনসেম্বল করার জন্য সবগুলো মডেলের একটি লিস্ট তৈরি
base_estimators = [
    ('LR', lr),
    ('DT', dt),
    ('SVM', svm),
    ('KNN', knn),
    ('GNB', gnb),
    ('XGBoost', xgb)
]

# ৩. এনসেম্বল মেথডসমূহ
hard_voting = VotingClassifier(estimators=base_estimators, voting='hard')
soft_voting = VotingClassifier(estimators=base_estimators, voting='soft')
stacking = StackingClassifier(estimators=base_estimators, final_estimator=LogisticRegression())

# সহজে লুপ চালানোর জন্য ডিকশনারি তৈরি
all_models = {
    'Logistic Regression (LR)': lr,
    'Decision Tree (DT)': dt,
    'SVM Classifier': svm,
    'KNN Classifier': knn,
    'Gaussian Naive Bayes (GNB)': gnb,
    'XGBoost Classifier': xgb,
    'Hard Voting Ensemble': hard_voting,
    'Soft Voting Ensemble': soft_voting,
    'Stacking Ensemble': stacking
}
```

```python
print("================ Training and Evaluating Models ================\n")

results = {}

for name, model in all_models.items():
    print(f"Training {name}...")
    model.fit(X_train, y_train) # মডেল ট্রেনিং

    # প্রেডিকশন এবং একিউরেসি চেক
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds) * 100
    results[name] = acc

    print(f"{name} Target Accuracy: {acc:.2f}%\n")

print("====================== FINAL RANKING ======================")
# একিউরেসি অনুযায়ী ক্রমানুসারে সাজিয়ে প্রিন্ট করা
sorted_results = sorted(results.items(), key=lambda x: x[1], reverse=True)
for rank, (model_name, accuracy) in enumerate(sorted_results, 1):
    print(f"Rank {rank}: {model_name} -> {accuracy:.2f}%")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image_dataset_from_directory

dataset = image_dataset_from_directory(
    '/content/drive/MyDrive/Rice_Disease_Project/',
    image_size=(224, 224),
    batch_size=32
)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# ১. ডাটা লোড করা এবং ট্রেন/ভ্যালিডেশন স্প্লিট করা
train_ds = image_dataset_from_directory(
    '/content/drive/MyDrive/Rice_Disease_Project/',
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(224, 224),
    batch_size=32
)

val_ds = image_dataset_from_directory(
    '/content/drive/MyDrive/Rice_Disease_Project/',
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(224, 224),
    batch_size=32
)

# ২. মডেল তৈরি করা
model = models.Sequential([
    layers.Rescaling(1./255, input_shape=(224, 224, 3)),
    layers.Conv2D(32, 3, activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(3, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# ৩. এখন মডেল ট্রেইন করা
model.fit(train_ds, validation_data=val_ds, epochs=10)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# ১. ভ্যালিডেশন ডাটা থেকে প্রকৃত লেবেল (y_true) এবং প্রেডিকশন (y_pred) বের করা
y_true = []
y_pred = []

for images, labels in val_ds:
    preds = model.predict(images)
    y_pred.extend(np.argmax(preds, axis=1))
    y_true.extend(labels.numpy())

# ২. কনফিউশন ম্যাট্রিক্স তৈরি
cm = confusion_matrix(y_true, y_pred)

# ৩. প্লট করা
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Bacterial', 'Brown', 'Leaf Smut'],
            yticklabels=['Bacterial', 'Brown', 'Leaf Smut'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
